```python
sea_level_avg_res=[]
for fileindex in range(initval, finalval):
  file=open(path_results,'a')
  filename=os.listdir(fn)[fileindex]
  filepath=fn+"/"+filename
  print(filepath)
  ds = nc.Dataset(filepath)
  #code 
  height_total=0;
  height_weighted=0;
  count_cells=0;
  area=0;
  sea_area=0;
  r=6371.0 # radius of earth in km
  jumps=2; # we will go through the different cells in jumps of 2 to shorten the running time
  latsize=ds.dimensions['latitude'].size # how many latitudes
  lonsize=ds.dimensions['longitude'].size # how many longitudes
  dtheta=np.pi/(latsize/jumps) # angular size of each latitude cell
  dphi=2*np.pi/(lonsize/jumps) # angular size of each longitude cell
  for i in range(269,451,2): # between -22.5 degrees and 22.5 degrees
    for j in range(int(lonsize/jumps)): # over all longitudes
      sla=ds.variables['sla'][0,i,jumps*j] # sea level anomaly of the specific cell
      theta=90+ds.variables['latitude'][i].data 
      area_element=r*r*np.sin(np.pi/180.0*theta)*dtheta*dphi # area of cell
      area=area+area_element
      if sla is not ma.masked:
        height_total=height_total+sla #total height - not very useful
        height_weighted=height_weighted+sla*area_element 
        # total sea level change weighted by area
        sea_area=sea_area+area_element # total area
        count_cells=count_cells+1
  print("--- %s area over area of sphere ---" %(area/(r*r*4*np.pi)))
  print("--- %s naive sea level anomaly average ---" %(100*height_total/count_cells))
  print("--- %s see level anomaly average weighted---" %(100*height_weighted/sea_area))
  sea_level_avg=100*height_weighted/sea_area  #(sla was given in m - changing it to cm)
  timeres=1950+ds.variables['time'][0].data/365 
  # which time is the file refering to (years from 1950)
  sea_level_avg_res= sea_level_avg_res+[[timeres,sea_level_avg]] #full list
  print("--- %s time since something---" %(timeres))
  # export to xls:
  file.write(str(fileindex)+","+filename+","+str(timeres)+","+str(sea_level_avg)+"\n")
  file.close()

print(sea_level_avg_res)
```
